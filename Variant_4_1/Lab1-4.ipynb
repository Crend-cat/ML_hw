### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

````python
import matplotlib.pyplot as plt
````

````python
data = pd.read_csv("russian-housing-market.csv")
````

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinski Zaton) с течением времени

````python
nagatino = data[data['sub_area'] == "Nagatinskij Zaton"].copy()
nagatino['timestamp'] = pd.to_datetime(nagatino['timestamp'])
nagatino['year_month'] = nagatino['timestamp'].dt.to_period('M')
prices = nagatino.groupby('year_month')['price_doc'].mean()

plt.figure(figsize=(14, 8))
plt.plot(prices.index.astype(str), prices.values, marker='o')
plt.title('Изменение средней цены квартиры в Нагатинском Затоне', fontsize=14, fontweight='bold')
plt.xlabel('Год - Месяц')
plt.ylabel('Средняя цена (руб)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
````

2. Постройте гистограмму распределения цен на квартиры с общей площадью более 50 кв.метров

````python
large_apartments = data[data.full_sq > 50]
plt.figure(figsize=(10, 9))
plt.hist(large_apartments['price_doc'], bins=101, edgecolor='black')
plt.title('Распределение цен на квартиры с общей площадью более 50 кв.метров', fontsize=14)
plt.xlabel('Цена (руб)')
plt.ylabel('Количество квартир')
plt.grid(True)
plt.tight_layout()
plt.show()
````

3. Визуализируйте количество проданных квартир в разных районах

````python
district_counts = data['sub_area'].value_counts()
plt.figure(figsize=(14, 8))
district_counts.head(20).plot(kind='bar')
plt.title('Количество проданных квартир по районам (Топ-20 районов)', fontsize=14, fontweight='bold')
plt.xlabel('Район')
plt.ylabel('Количество квартир')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

4. Постройте box plot отображающий распределение полных площадей квартир в Басманном районе

````python
basmannoe = data[data['sub_area'] == 'Basmannoe']

plt.figure(figsize=(10, 5))
plt.boxplot(basmannoe['full_sq'], patch_artist=True)
plt.title('Распределение полных площадей квартир в Басманном районе', fontsize=14, fontweight='bold')
plt.ylabel('Площадь кв.м')
plt.grid(True, alpha=0.3)
plt.show()
````

5. Визуализируйте зависимость средней стоимости квартир в зависимости от этажа

````python
floor_prices = data.groupby('floor')['price_doc'].mean()

plt.figure(figsize=(14, 7))
plt.plot(floor_prices.index, floor_prices.values, marker='o', linewidth=2, markersize=6)
plt.title('Зависимость средней стоимости квартир от этажа', fontsize=14, fontweight='bold')
plt.xlabel('Этаж')
plt.ylabel('Средняя цена (руб)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

6. Покажите как меняется средняя площадь квартир в зависимости от района

````python
district_areas = data.groupby('sub_area')['full_sq'].mean().sort_values(ascending=False)

plt.figure(figsize=(14, 8))
district_areas.head(20).plot(kind='bar')
plt.title('Топ-20 районов по средней площади квартир', fontsize=14, fontweight='bold')
plt.xlabel('Район')
plt.ylabel('Средняя площадь кв.м')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python
small_apartments = data[data['full_sq'] < 50].copy()
small_apartments['timestamp'] = pd.to_datetime(small_apartments['timestamp'])
small_apartments['year'] = small_apartments['timestamp'].dt.year
small_apartments_filtered = small_apartments[small_apartments['year'].between(2011, 2014)].copy()
small_apartments_filtered['year_month'] = small_apartments_filtered['timestamp'].dt.to_period('M')
monthly_median_prices = small_apartments_filtered.groupby('year_month')['price_doc'].median()

plt.figure(figsize=(14, 8))
plt.plot(monthly_median_prices.index.astype(str), monthly_median_prices.values, 
    marker='o', linewidth=2, markersize=4,)
plt.title('Изменение медианной цены на квартиру менее 50 кв.м за период с 2011 до 2014 года', fontsize=14, fontweight='bold')
plt.xlabel('Год-Месяц')
plt.ylabel('Медианная цена (руб)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````
